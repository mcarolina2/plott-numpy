# Plotagem De Gráficos  
* Meu primeiro contanto com a biblioteca Matplotlib. 
* Essa biblioteca é utilizada para a plotagem de gráficos simples.

```python
# importando as bibliotecas necessárias 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def teste(c):                            
    return c*c*c   

x = np.linspace(0, 3, 10)
y = teste(x)

plt.plot(x, y)       # Esse comando define que será um gráfico de linha
plt.plot(x, y, 'o')  #  Esse comando define que o ícone do gráfico será pontos
plt.show()           # Esse comando mostra o gráfico
```
